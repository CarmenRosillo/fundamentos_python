Edades clase 1

```python
#edades faltantes por clase

df_age_c1= df2[(df2[ "Pclass"]==1)& df2["Age"].isnull()]

# guardo las edades faltantes en la clase 1 sexo femenino 
df_age_c1_female= df_age_c1 [df_age_c1 ["Sex"] == "female"]

# guardo las edades faltantes en la clase 1 sexo masculino
df_age_c1_male= df_age_c1[df_age_c1["Sex"]== "male"]

```

```python
print (f"Edades faltantes clase 1, sexo femenino : {df_age_c1_female.shape[0]}") 
print (f"Edades faltantes clase 1, sexo masculino : {df_age_c1_male.shape[0]}") 
```

Edades clase 2

```python
#edades faltantes por clase

df_age_c2= df2[(df2[ "Pclass"]==2)& df2["Age"].isnull()]

# guardo las edades faltantes en la clase 1 sexo femenino 
df_age_c2_female= df_age_c2 [ df_age_c2 [ "Sex"]== "female"]

# guardo las edades faltantes en la clase 1 sexo masculino
df_age_c2_male= df_age_c2[df_age_c2["Sex"]== "male"]
```

```python
print (f"Edades faltantes clase 2, sexo femenino : {df_age_c2_female.shape[0]}")
print (f"Edades faltantes clase 2, sexo masculino : {df_age_c2_male.shape[0]}") 
```

clase 3

```python
#edades faltantes por clase

df_age_c3= df2[(df2[ "Pclass"]==3)& df2["Age"].isnull()]

# guardo las edades faltantes en la clase 1 sexo femenino 
df_age_c3_female= df_age_c3 [ df_age_c3 [ "Sex"]== "female"]

# guardo las edades faltantes en la clase 1 sexo masculino
df_age_c3_male= df_age_c3[df_age_c3["Sex"]== "male"]
```

```python
print (f"Edades faltantes clase 3, sexo femenino : {df_age_c3_female.shape[0]}")
print (f"Edades faltantes clase 3, sexo masculino : {df_age_c3_male.shape[0]}") 
```

```python
#Promedio de edad por clase y sexo
df3= df2.groupby(['Pclass', 'Sex'])['Age'].describe()
```

```python
df3
```

```python
#edades promedio por clase por sexo

c1_age_female= df3.loc[(1,"female"), "mean"]
c1_age_male= df3.loc[(1, "male"), "mean"]   

# clase 2
c2_age_female= df3.loc[(2, "female"), "mean"]
c2_age_male= df3.loc[(2, "male"), "mean"]

#clase 3
c3_age_female= df3.loc[(3, "female"), "mean"]   
c3_age_male= df3.loc[(3, "male"), "mean"]   
```

```python
print(c1_age_female)
print(c1_age_male)
print(c2_age_female)
print(c2_age_male)
print(c3_age_female)
print(c3_age_male)
```

### completar edades faltantes por clase por sexo

```python
df_age_c1_female["Age"]= df_age_c1_female ["Age"].fillna(round(c1_age_female,1))

df_age_c1_male["Age"]= df_age_c1_male ["Age"].fillna(round(c1_age_male,1))

df_age_c2_female["Age"]= df_age_c2_female ["Age"].fillna(round(c2_age_female,1))

df_age_c2_male["Age"]= df_age_c2_male ["Age"].fillna(round(c2_age_male,1))

df_age_c3_female["Age"]= df_age_c3_female ["Age"].fillna(round(c3_age_female,1))

df_age_c3_male["Age"]= df_age_c3_male ["Age"].fillna(round(c3_age_male,1))

```

```python
df_age_ok = pd.concat(
    [df_age_c1_female, df_age_c1_male, df_age_c2_female, 
     df_age_c2_male, df_age_c3_female, df_age_c3_male], ignore_index=True)
```
